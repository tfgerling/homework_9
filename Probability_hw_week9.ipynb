## 1 - Why normal data is an important consideration

Why is assuming normal data an important consideration when answering questions about probability?
There are many psychological and educational variables that are distributed approximately normally. In addition, it is easy for mathematical statisticians to work with. This means that many kinds of statistical tests can be derived for normal distributions. Statistical tests work very well even if the distribution is only approximately normally distributed. If the mean and standard deviation of a normal distribution are known, it is easy to convert back and forth from raw scores to percentiles.

## 2 - Histograms on Belize Airbnb data

The following code will create two histograms in a subplot showing the distribution of reviews_per_month and number_of_reviews in Belize Airbnb data

```python
# Setup 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats
from scipy.stats import norm

belize_df = pd.read_csv("belize_airbnb_listings.csv")


#create 2 histograms in a subplot

fig, ax = plt.subplots(1, 2, figsize=(15,5))

sns.histplot(ax=ax[0], x='reviews_per_month', data=belize_df, bins=5)
#set the title and labels for first plot (left)
ax[0].set_title('Histogram of Reviews per Month')
ax[0].set_ylabel('Number of Reviews')
ax[0].set_xlabel('Month')

sns.histplot(ax=ax[1], x='number_of_reviews', data=belize_df, bins=5)
#set the title and labels for second plot (right)
ax[1].set_title('Histogram of Number of Reviews')
ax[1].set_xlabel('Number of Reviews')

#set overall title 
fig.suptitle('Belize Airbnb Data')
plt.show()
```

### a. How do you choose your bin size? 

Depending on the actual data distribution and the goals of the analysis, different bin widths may be appropriate, so experimentation is usually needed to determine an appropriate width.Using wider bins where the density of the underlying data points is low reduces noise due to sampling randomness; using narrower bins where the density is high (so the signal drowns the noise) gives greater precision to the density estimation. I tried a number of different bins and looked at the data. I preferred fewer bins to make the columns wider and data easier to interpret.

# 6 - What percent of Airbnb listings in Belize are available 360 days a year or more?



In order to find the percentage that are available 360 days a year or more, the mean and standard deviation need to be calculated first. These values will be passed into the norm.cdf function from scipy. Since we want to find the number available 360 days a year or more, we subtract the value returned by norm.cdf from 1.

```python
# first calculate mean and standard deviation to use in calculating percentages
mean_avail = belize_df['availability_365'].mean()
std_avail = belize_df['availability_365'].std()

# pass mean and standard deviation into norm.cdf, subtracting the value returned from one to get the percent available
percent_avail_360_more = 1 - norm.cdf(360,mean_avail,std_avail)

print('Percent of Airbnb available 360 days or more: '+'{:.2%}'.format(percent_avail_360_more))
```

### a. Those available 360 days or less?


The same calculation can be used that was used above, except for subtracting from 1. This returns the ones available 360 days or less.

```python
# pass mean and standard deviation into norm.cdf to get the percent available
percent_avail_360_less = norm.cdf(360,mean_avail,std_avail)

print('Percent of Airbnb available 360 days or more: '+'{:.2%}'.format(percent_avail_360_less))
```

# 7 - What minimum night count are 75% of Airnbnb listing in Belize greater than?

```python
mean_min = belize_df['minimum_nights'].mean()
std_min = belize_df['minimum_nights'].std()

pct_75 = norm.ppf((1-0.75), mean_min, std_min)

print('Minimum night count 75% of Airbnb is greater than: '+'{:.2}'.format(pct_75))
```

### Monty Hall problem

The Monty Hall problem was named after the game show host Monty Hall. On his show (“Let’s Make a Deal”), contestants choose one of three doors with the option to switch after being shown what is behind a door they did not choose. Most people believe that switching does not matter because there are two unopened doors and one car (50/50 choice). The door opened depends on the player's initial choice, so the host is not opening a random door - assumption of independence does not hold. Before the host opens a door there is a 1/3 probability that the car is behind each door. If the car is behind door 1 the host can open either door 2 or door 3, so the probability that the car is behind door 1 and the host opens door 3 is 1/3 × 1/2 = 1/6. If the car is behind door 2 (and the player has picked door 1) the host must open door 3, so the probability that the car is behind door 2 and the host opens door 3 is 1/3 × 1 = 1/3. The key is that if the car is behind door 2 the host must open door 3, but if the car is behind door 1 the host can open either door.
